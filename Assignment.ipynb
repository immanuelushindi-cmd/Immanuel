```python
# inheritance
class Person:
    total_staff = 0

    def __init__(self, name, age, email):
        self.name = name
        self.age = age
        self.email = email
        Person.total_staff += 1

    def get_info(self):
        return f"Name: {self.name} | Age: {self.age} | Email: {self.email}"

    def __str__(self):
        return f"[Person] {self.name}, Age {self.age}"


class Student(Person):
    def __init__(self, name, age, email, scores):
        super().__init__(name, age, email)
        self.scores = scores

    def get_average(self):
        return sum(self.scores) / len(self.scores) if self.scores else 0

    def get_grade(self):
        avg = self.get_average()
        if avg >= 90:
            return "A"
        elif avg >= 80:
            return "B" 
        elif avg >= 70:
            return "C"
        elif avg >= 60:
            return "D"
        else:
            return "F"

    def __str__(self):
        return f"[Student] {self.name} | Avg: {self.get_average():.1f} | Grade: {self.get_grade()}"


class Teacher(Person):
    def __init__(self, name, age, email, subject, years_experience):
        super().__init__(name, age, email)
        self.subject = subject
        self.years_experience = years_experience

    def introduce(self):
        return f"Hi, I'm {self.name}. I teach {self.subject} and have {self.years_experience} years of experience."

    def evaluate_student(self, student):
        return f"{self.name} evaluated {student.name}: Avg = {student.get_average():.2f}, Grade = {student.get_grade()}"

    def __str__(self):
        return f"[Teacher] {self.name} | Subject: {self.subject} | Exp: {self.years_experience} yrs"


class HeadTeacher(Teacher):
    def __init__(self, name, age, email, subject, years_experience):
        super().__init__(name, age, email, subject, years_experience)
        self.managed_teachers = []

    def add_teacher(self, teacher):
        self.managed_teachers.append(teacher)

    def __str__(self):
        return f"[HeadTeacher] {self.name} | Managing: {len(self.managed_teachers)} teacher(s)"




p  = Person("George", 35, "george@mail.com")
s1 = Student("Alice", 20, "alice@gmail.com", [88, 92, 79, 95])
t1 = Teacher("Lancey", 40, "lancey@gmail.com", "Math", 15)
t2 = Teacher("Esdeekid", 38, "esdeekid@gmail.com", "Science", 20)
ht = HeadTeacher("Toliver", 50, "toliver@gmail.com", "Math", 30)
ht.add_teacher(t1)
ht.add_teacher(t2)

print(t1.introduce())

# 15

print(f"  HeadTeacher : {isinstance(ht, HeadTeacher)}")
print(f"  Teacher     : {isinstance(ht, Teacher)}")
print(f"  Person      : {isinstance(ht, Person)}")

# 16
print("\nMixed list")
for obj in [p, s1, t1, ht]:
    print(f"  {obj.get_info()}")

# B4
print(f"\nTotal staff created: {Person.total_staff}")

# B5

for obj in [p, s1, t1, ht]:
    print(f"  {obj}")

# B6
print(f"\n{t1.evaluate_student(s1)}")
```

```python
#Polymorphism
class Chart:
    def __init__(self, title, data):
        self.title = title
        self.data = data
        self.validate_data()  # B9

    def render(self):
        return "Rendering a generic chart"   #25

    def validate_data(self):  # B9
        if not isinstance(self.data, list) or len(self.data) == 0:
            raise ValueError("Data must be a non-empty list.")

    def __str__(self):  # B10
        return f"[Chart] {self.title} | Data points: {len(self.data)}"


class BarChart(Chart):  #26
    def render(self):
        return f"Bar Chart: '{self.title}' | Bars: {self.data}"

    def __str__(self):  # B10
        return f"[BarChart] {self.title} | Data points: {len(self.data)}"


class LineChart(Chart):  #27
    def render(self):
        return f"Line Chart: '{self.title}' | Points: {self.data}"

    def __str__(self):  # B10
        return f"[LineChart] {self.title} | Data points: {len(self.data)}"


class PieChart(Chart):  #28
    def render(self):
        return f"Pie Chart: '{self.title}' | Slices: {self.data}"

    def __str__(self):  # B10
        return f"[PieChart] {self.title} | Data points: {len(self.data)}"


class ScatterPlot:  # does NOT inherit from Chart - duck typing
    def __init__(self, title, data):
        self.title = title
        self.data = data

    def render(self):
        return f"Scatter Plot: '{self.title}' | Points: {self.data}"

    def __str__(self):  # B10
        return f"[ScatterPlot] {self.title} | Data points: {len(self.data)}"


def render_all(charts):
    for chart in charts:
        print(chart.render())


# Task 30 - dashboard
dashboard = [
    BarChart("Monthly Sales", [100, 200, 150, 300]),
    LineChart("Revenue Trend", [500, 600, 550, 700]),
    PieChart("Market Share", [40, 30, 20, 10]),
    ScatterPlot("Age vs Score", [(20, 88), (22, 92), (21, 79)]),
]

# Task 32
render_all(dashboard)

# B9
try:
    BarChart("Empty", [])
except ValueError as e:
    print(f"ValueError: {e}\n")

# B10
print("DASHBOARD\n")
for chart in dashboard:
    print(chart)
```

```python
class Person:
    total_registered = 0

    def __init__(self, name, age, national_id):
        self.name = name
        self.age = age
        self.national_id = national_id
        Person.total_registered += 1

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if not isinstance(value, int):
            raise TypeError("Age must be an integer.")
        if value < 0:
            raise ValueError("Age cannot be negative.")
        self._age = value

    @staticmethod
    def is_valid_id(id_number):
        return isinstance(id_number, str) and id_number.isdigit() and len(id_number) == 8

    def __eq__(self, other):
        if not isinstance(other, Person):
            return NotImplemented
        return self.national_id == other.national_id

    def __str__(self):
        return f"[Person] {self.name} | Age: {self.age} | ID: {self.national_id}"

    def introduce(self):
        return f"Name: {self.name} | Age: {self.age} | ID: {self.national_id}"


class Patient(Person):
    def __init__(self, name, age, national_id, temperature,
                 bp_systolic, bp_diastolic, weight_kg, height_cm):
        super().__init__(name, age, national_id)
        self.__medical_history = []
        self.__record_count = 0
        self._temperature = temperature
        self._bp_systolic = bp_systolic
        self._bp_diastolic = bp_diastolic
        self.weight_kg = weight_kg
        self._height_cm = height_cm
        self._emergency = False

    @property
    def weight_kg(self):
        return self._weight_kg

    @weight_kg.setter
    def weight_kg(self, value):
        if not (1 <= value <= 300):
            raise ValueError("Weight must be between 1 and 300 kg.")
        self._weight_kg = value

    @property
    def bmi(self):
        height_m = self._height_cm / 100
        return round(self._weight_kg / (height_m ** 2), 2)

    @property
    def triage_priority(self):
        if self._emergency:
            return 0

        score = 0

        temp = self._temperature
        if temp >= 40 or temp < 35:
            score += 1
        elif temp >= 38.5 or temp < 36:
            score += 2
        else:
            score += 3

        s, d = self._bp_systolic, self._bp_diastolic
        if s >= 180 or d >= 120 or s < 80:
            score += 1
        elif s >= 140 or d >= 90:
            score += 2
        else:
            score += 3

        a = self._age
        if a < 5 or a > 70:
            score += 1
        elif a < 13 or a > 60:
            score += 2
        else:
            score += 3

        return score

    @property
    def medical_history(self):
        return self.__medical_history.copy()

    def add_record(self, note):
        if not isinstance(note, str) or len(note) == 0:
            raise ValueError("Note must be a non-empty string.")
        self.__record_count += 1
        self.__medical_history.append(f"[Entry {self.__record_count}] {note}")

    @classmethod
    def from_dict(cls, d):
        return cls(
            d["name"], d["age"], d["national_id"],
            d["temperature"], d["bp_systolic"], d["bp_diastolic"],
            d["weight_kg"], d["height_cm"],
        )

    def flag_emergency(self):
        self._emergency = True

    def __lt__(self, other):
        return self.triage_priority < other.triage_priority

    def __str__(self):
        return f"[Patient] {self.name} | BMI: {self.bmi} | Score: {self.triage_priority}"

    def introduce(self):
        base = super().introduce()
        return f"{base} | BMI: {self.bmi} | Triage Score: {self.triage_priority}"


class MedicalStaff(Person):
    def __init__(self, name, age, national_id, staff_number):
        super().__init__(name, age, national_id)
        self.staff_number = staff_number

    def attend(self, patient):
        raise NotImplementedError

    def __str__(self):
        return f"[MedicalStaff] {self.name} | Staff No: {self.staff_number}"

    def introduce(self):
        base = super().introduce()
        return f"{base} | Staff No: {self.staff_number}"


class Prescription:
    def __init__(self, patient, medication, dosage, doctor):
        self.patient_name = patient.name
        self.medication = medication
        self.dosage = dosage
        self.doctor_name = doctor.name

    def __str__(self):
        return (
            f"     Rx | {self.patient_name} | "
            f"{self.medication} {self.dosage} | "
            f"Dr. {self.doctor_name}"
        )


class Nurse(MedicalStaff):
    def __init__(self, name, age, national_id, staff_number):
        super().__init__(name, age, national_id, staff_number)

    def __str__(self):
        return f"[Nurse] {self.name} | Staff No: {self.staff_number}"

    def introduce(self):
        base = super().introduce()
        return f"{base} | Role: Nurse"

    def attend(self, patient):
        patient.add_record(f"Nurse {self.name}: vitals recorded and triage updated.")
        print(f"  [Nurse    ] {self.name} attended {patient.name}.")


class Doctor(MedicalStaff):
    def __init__(self, name, age, national_id, staff_number):
        super().__init__(name, age, national_id, staff_number)

    def __str__(self):
        return f"[Doctor] {self.name} | Staff No: {self.staff_number}"

    def introduce(self):
        base = super().introduce()
        return f"{base} | Role: Doctor"

    def attend(self, patient):
        rx = Prescription(patient, "Amoxicillin", "500mg TID x 5 days", self)
        patient.add_record(f"Dr. {self.name}: examined patient, prescription issued.")
        print(f"  [Doctor   ] Dr. {self.name} examined {patient.name}.")
        return rx


class LabTechnician:
    def __init__(self, name, lab_id):
        self.name = name
        self.lab_id = lab_id

    def attend(self, patient):
        patient.add_record(f"Lab Tech {self.name}: samples collected.")
        print(f"  [Lab Tech ] {self.name} collected samples from {patient.name}.")

    def __str__(self):
        return f"[LabTechnician] {self.name} | Lab ID: {self.lab_id}"


class TriageQueue:
    def __init__(self):
        self._queue = []

    def add_patient(self, patient):
        self._queue.append(patient)
        self._queue.sort()

    def emergency_override(self, patient):
        patient.flag_emergency()
        if patient not in self._queue:
            self._queue.append(patient)
        self._queue.sort()
        print(f"EMERGENCY: {patient.name} flagged critical — queue re-sorted.")

    def __len__(self):
        return len(self._queue)

    def __iter__(self):
        return iter(self._queue)

    def __contains__(self, patient):
        return patient in self._queue

    def __getitem__(self, index):
        return self._queue[index]

    def __str__(self):
        rows = [
            f"  {i + 1}. {p.name:<22} Score: {p.triage_priority}"
            for i, p in enumerate(self._queue)
        ]
        return "\n".join(rows)


if __name__ == "__main__":
    SEP = "=" * 65

    print(SEP)
    print("  NAIROBI CLINIC — TRIAGE & PATIENT QUEUE SYSTEM")
    print(SEP)

    # --- Registration ---
    p1 = Patient.from_dict({
        "name": "Amina Wanjiru", "age": 72, "national_id": "12345678",
        "temperature": 40.5, "bp_systolic": 190, "bp_diastolic": 125,
        "weight_kg": 58, "height_cm": 160,
    })
    p2 = Patient("Brian Otieno",    30, "87654321", 36.8, 120, 80, 75, 175)
    p3 = Patient("Cynthia Muthoni",  3, "11223344", 39.1,  95, 65, 14,  95)

    nurse    = Nurse("Sarah Kamau",      35, "55667788", "N-001")
    doctor   = Doctor("James Odhiambo", 45, "99887766", "D-001")
    lab_tech = LabTechnician("Kevin Njoroge", "LT-007")

    print("\n--- Patient Registration ---")
    for p in [p1, p2, p3]:
        valid = Person.is_valid_id(p.national_id)
        print(f"  {p.name:<22} ID: {p.national_id}  Valid: {valid}")

    print("\n--- Staff on Duty ---")
    for staff in [nurse, doctor]:
        print(f"  {staff.introduce()}")
    print(f"  {lab_tech}")

    print(f"\n  Total persons registered: {Person.total_registered}")
    print(f"  (Patients + Medical Staff only. LabTechnician is not a Person.)")

    print("\n--- ID Validation ---")
    for id_ in ["12345678", "1234", "ABCD1234"]:
        print(f"  {id_:<12} → {'Valid' if Person.is_valid_id(id_) else 'Invalid'}")

    print("\n--- Waiting Queue (sorted by urgency) ---")
    queue = TriageQueue()
    queue.add_patient(p2)
    queue.add_patient(p1)
    queue.add_patient(p3)
    print(queue)
    print(f"\n  Patients waiting : {len(queue)}")
    print(f"  Next to be seen  : {queue[0].name}")
    print(f"  Amina in queue   : {p1 in queue}")

    print("\n--- Treatment Order ---")
    for i, p in enumerate(sorted(queue), 1):
        print(f"  {i}. {p.name:<22} Triage Score: {p.triage_priority}  BMI: {p.bmi}")

    print(f"\n--- Attending to {p1.name} ---")
    for staff in [nurse, doctor, lab_tech]:
        result = staff.attend(p1)
        if isinstance(result, Prescription):
            print(result)

    print(f"\n--- Medical History: {p1.name} ---")
    for record in p1.medical_history:
        print(f"  {record}")

    print("\n--- Staff Type Checks ---")
    for person, label in [(doctor, "Doctor"), (lab_tech, "Lab Tech")]:
        print(f"  {label:<10} is MedicalStaff: {isinstance(person, MedicalStaff)}"
              f"  |  is Person: {isinstance(person, Person)}")

    print("\n--- Duplicate Patient Check ---")
    p_dup = Patient("Amina W.", 72, "12345678", 36.8, 120, 80, 58, 160)
    print(f"  Same ID as Amina Wanjiru?  {p1 == p_dup}")
    print(f"  Same ID as Brian Otieno?   {p1 == p2}")

    print("\n--- Emergency Admission ---")
    critical = Patient("David Kipchoge", 2, "44332211", 41.5, 60, 30, 12, 82)
    queue.add_patient(critical)
    print("  Queue before emergency:")
    print(queue)
    print()
    queue.emergency_override(critical)
    print("  Queue after emergency:")
    print(queue)

    print("\n--- Input Validation ---")
    tests = [
        ("Age = -5",          lambda: setattr(p2, "age", -5)),
        ("Age = 'thirty'",    lambda: setattr(p2, "age", "thirty")),
        ("Weight = 400 kg",   lambda: setattr(p2, "weight_kg", 400)),
        ("Weight = 0 kg",     lambda: setattr(p2, "weight_kg", 0)),
        ("Empty note",        lambda: p1.add_record("")),
    ]
    for label, call in tests:
        try:
            call()
        except (TypeError, ValueError) as e:
            print(f"  {label:<18} → {type(e).__name__}: {e}")

    print("\n--- Patient Summary ---")
    for p in [p1, p2, p3, critical]:
        tag = " [CRITICAL — Emergency]" if p._emergency else ""
        print(f"  {p.name:<22} BMI: {p.bmi:<8} Triage Score: {p.triage_priority}{tag}")

    print(f"\n  Total registered: {Person.total_registered}")
    print(SEP)
```

```python
# Prior probabilities
p_sick = 1 / 1000
p_healthy = 1 - p_sick

# Test accuracies
p_pos_given_sick = 0.95
p_pos_given_healthy = 0.05 

# Total probability of testing positive
p_positive = (p_pos_given_sick * p_sick) + (p_pos_given_healthy * p_healthy)

# Bayes' Theorem
p_sick_given_pos = (p_pos_given_sick * p_sick) / p_positive

print(f"Chance of actually being sick: {p_sick_given_pos:.4f}")
```

```python
seed = 42

def next_random():
    global seed
    seed = (1103515245 * seed + 12345) % (2**31)
    return seed / (2**31)

def sqrt(x):
    guess = x / 2
    for _ in range(25):
        guess = (guess + x / guess) / 2
    return guess

def mean(values):
    return sum(values) / len(values)

def std_dev(values):
    avg = mean(values)
    return sqrt(mean([(v - avg) ** 2 for v in values]))

def skewness(values):
    avg = mean(values)
    sd = std_dev(values)
    return mean([(v - avg) ** 3 for v in values]) / (sd ** 3)

def skewed_value():
    u = next_random()
    return (u ** 3) * 500   # cubing a uniform number bunches values near 0, with a long tail

def print_histogram(values, title, bins=20, width=50):
    low = min(values)
    high = max(values)
    step = (high - low) / bins

    counts = [0] * bins
    for v in values:
        i = int((v - low) / step)
        if i == bins:
            i = bins - 1
        counts[i] += 1

    biggest = max(counts)
    print("\n" + title)
    for i in range(bins):
        bar = "#" * int((counts[i] / biggest) * width)
        print(f"{low + i * step:7.1f} | {bar} ({counts[i]})")


population = [skewed_value() for _ in range(20000)]

sample_means = []
for _ in range(2000):
    sample = [population[int(next_random() * len(population))] for _ in range(30)]
    sample_means.append(mean(sample))

print(f"Population mean: {mean(population):.2f}")
print(f"Population skewness: {skewness(population):.2f}")
print(f"Sample means mean: {mean(sample_means):.2f}")
print(f"Sample means skewness: {skewness(sample_means):.2f}")

print_histogram(population, "Original population (skewed)")
print_histogram(sample_means, "Sample averages (should look bell-shaped)")
```
