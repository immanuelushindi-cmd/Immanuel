SET 
A set is an UNORDRED collection with NO DUPLICTES
Sets come from mathematics - they represent a well- defined group of distinct objects.
Because sets have no order, you CANNOT access items by index
(no my set[0]) 
Sets use curly braces {}. An EMPTY set must be writtrn as set() - not {} (that creates a dictionary).
Main uses: remiving duplicates from data, and performing set theory operations


```python
#SET
#Creating a Set
score_set = {85,92,78,44,55,63}
# Converting a list to a set
raw_scores = [85,92,78,44,55,63]
unique_score = set(raw_scores)
print(unique_score)
```

```python
class_a = {85,92,78}
class_b = {78,65,92,55}
#Union- combines all unique elements from both sets
union = class_a | class_b
print(union)

#Intersection- gives you only the elements that are present in both sets
intersection = class_a & class_b
print(intersection)

#Difference-gives you the elements that are present in one set but not the other
difference = class_a - class_b
print(difference)

#Symmetric Difference- gives you the elements that are present in either set but not both
sym_diff = class_a ^ class_b
print(sym_diff)

print()

#USING METHODS
union1 = class_a.union(class_b)
print(union1)

intersection1 = class_a.intersection(class_b)
print(intersection1)

difference1 = class_a.difference(class_b)
print(difference1)

sym_diff1 = class_a.symmetric_difference(class_b)
print(sym_diff1)
```

```python
class_a = {85,92,78}
class_b = {78,65,92,55}

#UNION - everything in A or B
union = class_a | class_b
union1 = class_a.union(class_b)
print("union:", union)

#INTERSECTION - only in BOTH
intersection = class_a & class_b
intersection1 = class_a.intersection(class_b)
print("intersection:", intersection)

#DIFFERENCE - in A but NOT in B
difference = class_a - class_b
difference1 = class_a.difference(class_b)
print("difference:", difference)

#SYMMETRIC DIFFERENCE - in A or B but NOT BOTH
sym_diff = class_a ^ class_b
sym_diff1 = class_a.symmetric_difference(class_b)
print("symmetric difference:", sym_diff)
```

```python
# Default Parameters
# A parameter can have a DEFAULT VALUE used when the caller does not provide an argument for that parameter. This allows the function to be called with fewer arguments than it is defined to accept, while still providing reasonable default behavior.
# Parameters with default values must come after parameters without default values in the function definition. This is because Python matches arguments to parameters based on their position, and having a parameter with a default value before one without would create ambiguity in how arguments are assigned to parameters.
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"
print(greet("Immanuel"))  # Output: Hello, Immanuel!- this uses the default greeting "Hello"
print(greet("Immanuel", "Hi"))  # Output: Hi, Immanuel!

# ARGUMENTS: the actual value or object passed into a function or method when you call it. 
```

```python
# Exception Handling
# try  - put the code that MIGHT fail here
# except - handle the exception if it occurs
# else - execute this block if no exception occurs
# finally - execute this block regardless of whether an exception occurred or not

def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero!"
    except TypeError:
        return "Error: Both arguments must be numbers!"
    else:
        return result
    finally:
        print("Division attempt completed.")

print(safe_divide(10, 2))
print(safe_divide(10, 0))  
print(safe_divide(10, "a"))
```

```python
# INHERTANCE
# Parent Blueprint
class Person:
    def __init__(self, name, city):
        self.name = name
        self.city = city

# Child Blueprint (inherits from Person)
class Student(Person):
    def __init__(self, name, city, student_id):
        super().__init__(name, city) # Sends name and city to the Parent
        self.student_id = student_id  # Unique to Student

# Student automatically has access to name and city
student1 = Student("Immanuel Ushindi", "Nairobi", "STU123")
print(student1.name)  
```

```python
# POLYMORPHISM
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        return f"I am a learner named {self.name}."

class Teacher:
    def __init__(self, name, subject):
        self.name = name
        self.subject = subject

    def introduce(self):
        return f"I am an instructor teaching {self.subject}."

# A list holding two different types of objects
classroom = [Student("Immanuel", 19), Teacher("Smith", "Math")]

# One loop triggers two different responses
for person in classroom:
    print(person.introduce())
    
```

```python
# Abstraction
from abc import ABC, abstractmethod

# The Master Rules Template
class SchoolMember(ABC):
    @abstractmethod
    def get_schedule(self):
        pass  # Just a rule saying: "You MUST create this action later!"

# This class follows the rule
class Student(SchoolMember):
    def get_schedule(self):
        return "Classes from 8 AM to 3 PM."
    
print(Student().get_schedule())
```

## Encapsulation
- It can mean 2 things in OOP
    - **Bundling** Keeping data(attributes) and the methods(behaviour) that operate data together in one class.
    This keeps related code organised and prevents scattered, diconnected logic
    - **Access Control**: Controlling who can see an modify an objects's internal data.
    Some data can be changed only by specific validated methods - not directly from outside the class
#### Types of attributes in encapsulation
- **PUBLIC**(name): Accessibe from anywhere - inside the class, outside it, by subclasses
- **PROTECTED**(_name): Intended for inernal use only - not accessible from outside the class
- **PRIVATED**(__name): This makes accidental access from outside the class very difficult


```python
#Encapsulation
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self._bank = "Equitybank"
        self.__balance = balance
    
    def get_balance(self):
        return self.__balance
    
    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit must be positive")
        self.__balance += amount
        return f"Deposited {amount:.2f}. New balance: {self.__balance:.2f}"
    
    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive")
        if amount > self.__balance:
            raise ValueError("Insufficient funds")
        self.__balance -= amount
        return f"Withdrawn {amount:.2f}. New balance: {self.__balance:.2f}"
    
acc = BankAccount("Immanuel", 1000)
print(acc.owner)
print(acc.get_balance())
print(acc.deposit(500))
print(acc.withdraw(200))

print(acc._BankAccount__balance)
```

##DECORATORS AND PROPERTIES
###Decorator
- It is a function-upgrader
- A function that takes a functionin and returns an improved version
- Denoted using "@"
- POLYMORPHISM

```python
def shout(x):  # Takes a function in
    def wrapper():
        return x().upper() + "!!"
    return wrapper                # Returns the upgraded function

@shout              # Shorthand for greet = shout(greet)
def greet():
    return "Hello"  # HELLO !! :greet was replaced by the upgraded version
```

### Property
Is a fake attribute
It looks like a variable from the outside but secretly has a fuction running on the inside
A normal attribute is like a box: reading it fetches the stored value.
A **PROPERTY** has NO box: it runs a getter and a setter function when you read or write to it.

## POLYMORPHISM
-This means that different classes can share the same method, name but each class implements it differently.
-You can call the same method on objects of different types and the right version runs automatically - no if/else needed
-It is powerful because it lets you write generic code that works with objects of any class as long as they have the right method - you do not need to know or check what type of object you are dealing with.


```python
# POLYMORPHISM
class Animal:
    def __init__(self, name):
        self.name = name
    def speak(self):
        return f"{self.name} makes a sound"
    def describe(self):
        return f"I am a {self.name}, {type(self).__name__}"
    
class Dog(Animal):
    def speak(self):          # OVERIDES Animal.speak()
        return f"{self.name} says WOOF!"

class Cat(Animal):
    def speak(self):          # OVERIDES Animal.speak()
        return f"{self.name} says meow!"
    
class Duck(Animal):
    def speak(self):          # OVERIDES Animal.speak()
        return f"{self.name} says Quack!"

# POLYMORPHISM
animals = [Dog("Rex"), Cat("Whiskers"), Duck("Donald"), Animal("Generic")]
# Same loop, same method call -  each object does the right thing automatically
for animal in animals:
    print(animal.speak())  # each calls ITS OWN version

for animal in animals:
    print(animal.describe())  # each calls ITS OWN version


# describe() is not overidden - all objects use the same version
```

## Duck Typing
- Two classes do not need to share a parent class to be used polymorphically, they just need to share the same the same method name
- An object suitability for a task is determined by the presence of the specific method and attribute rather than actual class type or inheritance hierarchy

```python
# These classes do not inherit from the same parent
#  But they all have a summarise() method  - that is enough

class ExamResult:
    def __init__(self,subject,score):
        self.subject = subject
        self.score = score

    def summarise(self):
        return f"Exam: {self.subject} - Score: {self.score}/100"
    
class Project:
    def __init__(self, title, grade):
        self.title = title
        self.grade = grade
    
    def summarise(self):
        return f"Project: {self.title} - Grade: {self.grade}"
        
    

class Attendance:
    def __init__(self, days_present, days_total):
        self.days_present = days_present
        self.days_total = days_total
    
    def summarise(self):
        pct = (self.days_present / self.days_total) * 100
        return f"Attendance: {self.days_present}/{self.days_total} ({pct:.1f}%)"
    
exams = [ExamResult("Maths",50),ExamResult("Physics",60),ExamResult("Chemistry",70)]
projects = [Project("Project 1",80),Project("Project 2",90),Project("Project 3",95)]
attendance = [Attendance(10,20),Attendance(15,20),Attendance(20,20)]

for exam in exams:
    print(exam.summarise())
    
for project in projects:
    print(project.summarise())

for attendance in attendance:
    print(attendance.summarise())
```
