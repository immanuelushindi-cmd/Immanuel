### Figures and Axes - The CANVAS
- In matplotlib, the canvas has 2 parts:
     - the figure: the entire image
     - the axes individual plots within the figure

### plt.figure() - Creating the canvas

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
print("ready")
```

```python
fig = plt.figure(figsize=(10, 6), dpi=100, facecolor='white', edgecolor='black', linewidth=2)
```

### plt.subplots() - Multiplication

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,10), sharex=False, sharey=False)
```

### Line Plot
- Shoes change over a continuous variable (usually time). Trends, patterns, cycles.
- When to use: Time series data, showing progression

```python
x=np.arange(1, 11)
y=np.array([10, 15, 20, 25, 30, 35, 40, 45, 50, 55])
plt.plot(x, y,
         color='blue', 
         linestyle='-',
         linewidth=2,
         marker='o',
         markersize=6,
         markerfacecolor='white',
         markeredgecolor='blue',
         alpha=0.8,
         label='Revenue 2024',
         zorder=3,
         )
```
