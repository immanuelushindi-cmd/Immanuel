## NaN section and the rest of numpy assignment

```python
daily_sales_float = daily_sales.astype(float)

daily_sales_float[2, 1] = np.nan
daily_sales_float[5, 3] = np.nan
daily_sales_float[8, 0] = np.nan

print(daily_sales_float)
```

```python
missing = np.isnan(daily_sales_float)

print(missing)
print(missing.sum())
print(missing.sum(axis=0))
```

```python
nan_means = np.nanmean(daily_sales_float, axis=0)
print(nan_means)

regular_means = daily_sales_float.mean(axis=0)
print(regular_means)
```

```python
col_medians = np.nanmedian(daily_sales_float, axis=0)
print(col_medians)

filled = daily_sales_float.copy()
row_idx, col_idx = np.where(np.isnan(filled))
filled[row_idx, col_idx] = col_medians[col_idx]

print(filled)
print(np.isnan(filled).sum())
```

```python
def clean_array(arr, sentinel=-1):
    arr = arr.astype(float)
    arr = np.where(arr == sentinel, np.nan, arr)

    col_means = np.nanmean(arr, axis=0)
    row_idx, col_idx = np.where(np.isnan(arr))
    arr[row_idx, col_idx] = col_means[col_idx]

    return arr

test_array = np.array([[12000, -1, 8000], [11000, 9500, -1], [-1, 10500, 7800]])
print(clean_array(test_array))
```
