```python
import numpy as np
print("ready")
```

```python
np.random.seed(1)

categories = ["Food", "Electronics", "Clothing", "Furniture"]
day_numbers = np.arange(1, 11)

daily_sales = np.random.randint(8000, 65000, size=(10, 4))
units_sold = np.random.randint(15, 350, size=(10, 4))

print(daily_sales)

print()

print(units_sold)
```

```python
print(daily_sales.shape)
print(daily_sales.ndim)
print(daily_sales.size)
print(daily_sales.dtype)

print()

print(units_sold.shape)
print(units_sold.ndim)
print(units_sold.size)
print(units_sold.dtype)
```

```python
targets = np.full((10, 4), 15000)
print(targets)
print(targets.shape)
print(targets.ndim)
print(targets.dtype)
```

```python
spaced = np.linspace(daily_sales.min(), daily_sales.max(), 10)
print(spaced)
```

```python
np.random.seed(7)
clothing = np.random.normal(9500, 1200, 10)
print(clothing)
print(clothing.mean())
print(daily_sales[:, 2].mean())
```

```python
electronics = daily_sales[:, 1]
print(electronics)
print(electronics.max())
```

```python
last_3_food = daily_sales[-3:, 0]
print(last_3_food)
print(last_3_food.mean())
```

```python
cheap_cols = np.arange(0, 4, 2)
print(cheap_cols)

subgrid = daily_sales[:5, cheap_cols]
print(subgrid)
print(subgrid.shape)
```

```python
even_days = daily_sales[::2, :]
print(even_days)
print(even_days.sum(axis=0))
```

```python
sub = daily_sales[3:6, 1:3].copy()
sub += 500
print(sub)
print(daily_sales[3:6, 1:3])
```

```python
sales_with_vat = daily_sales * 1.16
print(sales_with_vat)
```

```python
revenue_per_unit = (daily_sales / units_sold).round(2)
print(revenue_per_unit)

avg_per_unit = revenue_per_unit.mean(axis=0)
print(avg_per_unit)
print(categories[avg_per_unit.argmax()])
```

```python
diff = daily_sales - 15000
print(diff)

below_target = (diff < 0).sum(axis=0)
print(below_target)
# order is Food, Electronics, Clothing, Furniture
```

```python
log_sales = np.log(daily_sales)
print(log_sales)
print(log_sales.mean())
```

```python
weighted_sales = daily_sales * np.array([1.0, 0.9, 1.0, 0.95])
print(weighted_sales)
```

```python
category_avg = daily_sales.mean(axis=0)
print(category_avg)
print(categories[category_avg.argmax()])
print(category_avg[category_avg.argmax()])
```

```python
daily_totals = daily_sales.sum(axis=1)
print(daily_totals)
print(day_numbers[daily_totals.argmax()])
print(day_numbers[daily_totals.argmin()])
```

```python
category_std = daily_sales.std(axis=0)
print(category_std)
print(categories[category_std.argmax()])
```

```python
cum_electronics = daily_sales[:, 1].cumsum()
print(cum_electronics)

crossed = cum_electronics > 400000
print(day_numbers[crossed.argmax()])
```

```python
electronics_share = daily_sales[:, 1] / daily_sales.sum(axis=1) * 100
print(electronics_share.round(2))
print(electronics_share.mean())
```
