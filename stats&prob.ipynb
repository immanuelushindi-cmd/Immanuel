### STATISTICS
#### What is Statistics
- It is the science of collecting, organising, summarising, analysing and interpreting data in order to make decisions under uncertainty.

#### Branches of Statistics
- **Descriptive**: It answers the question: "What does the data I have look like?" e.g: The Average of fare of 40 surveyed taxis was 100.
- **Inferential**: It answers the question: "What can I conclude about the data I don't have?"
e.g: Therefore the average fare across all of Nairobi taxis is probably between 66 - 74

#### Types of Data
- **QUALITATIVE (categorical)**
    -***Norminal***: Labels with no oreder e.g Network: safaricom, airtel etc
    -***Ordinal***: Ordered labels, unequal gaps e.g:
    Satisfaction level-low/medium/high
-**QUANTITATIVE (numerical)**: 
    -***Discrete***: Countable whole numbers e.g number of passengers: 1,2,3...
    -***Continuous***: Any value in a range e.g Trip distance: 12.56km

#### Population 
- A population is every unit you care about.
- Is  the subset you actually observe

- We sample because measuring the whole population is usually impossible, too expensive or too slow
- The art of inferential statistics is making the sample representative so that conclusions drwan from it generalise.

#### Parameter
- It is a number describing the population (μ,σ)

#### A statistic
- Is a number describing a sample (x,s)
- We compute statistics to estimate parameters

#### Sampling methods
- Simple random sampling
- Systematic sampling
- Stratified sampling
- Cluster sampling
- Convenience sampling

### Hypothesis Testing
-It is a formal procedure for deciding whether a pattern in our SAMPLE is real enough to believe about the population, or whether it could easily be an error of random sampling.
-Everytime we ask "is this difference significant" question i.e. does the new ,atatu earn more?; does the variant  B concert better? - is a hypothesis test

### Central Limit Theorem
- It says: if you take any samples and compute each one's mean, the distribution of those samples means is approximately mormal regardless of the shape of the original data -  as long as the sample size is large enough.

#### Intuition
- Individual maatu fares might be widely skewed (thst 250 outlier!). But the average fare of 40 trips is a much steadier thing - extreme highs and lows tend to cancel within each sample. Do that averaging many times and the averages pile up into a neat bell curve centred on the true population mean. The CLT is why the distribution shows up everywhere in inference even when the raw data isn't normal

### Standard Deviation and Standard Error
Std DEV(σ) = How spread out the individual data points are
Std Error(σ,/√n) = How spread out the sample mean is across repeated samples
