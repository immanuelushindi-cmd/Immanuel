## Accessing and Selecting Data

```python
# Selecting rows : loc and iloc
# loc - use labels (row index value, column names). Slice end is included
# iloc - use positions(numbers like a list). Slice end is exclusive

df.iloc[0]
df.iloc[0:3]
df.iloc[0:5, 0:3] 
df.loc[0] # row label 0
df.loc[0:4] # labels 0 to 4 inclusive -5 row
df.loc[0:4, 'name':'city'] # prints columns from name to city inclusive

```

```python
df.loc[2:5, ['name', 'salary_kes']]
```

```python
df.iloc[:10, [0,2,3]] # position
```

```python
# Employees earning over 100000
df[df['salary'] > 100000]

# Data Analysts only
df[df['job_title'] == 'Data Analyst']

# Nairobi employees earning over 100000
df[(df['city'] == 'Nairobi') & (df['salary'] > 100000)]

# Data Analysts or ML Engineers
df[(df['job_title'] == 'Data Analyst') | (df['job_title'] == 'ML Engineer')]

# Everyone who is not on contract
df[~(df['employment_type'] == 'Contract')]
```

```python
df[df['job_title'].isin(['Data Analyst', 'ML Engineer'])]
```

```python
# Detecting the missing value
# Missing values in pandas show up as NaN(Not a Number)
# We find them first thendecide what to do with them
df.isna()
df.isna().sum()
df['salary_history_2022'].isna()

# Filter rows where values are missing
df[df['salary_history_2022'].notna()]
# 
```

```python
df.drop() 
df.drop(how='all') 
df.drop(subset=['salary_history_2022']) 
df.drop(axis=1)
df.drop(thresh=8)
```

```python
# Filling in Missing Values
df.fillna(0)

# Fill with the median of that column
median_val = df['salary_history_2022'].median()
df['salary_history_2022'].fillna(median_val, inplace=True)

# Fill text column with the most common value
df
```

```python
# Interpolation
# For numeric columns in sequence, estimate missing values by interpolation
df['salary_history_2022'].interpolate(method='linear', inplace=True)

# Flagging missing values
# Used when you want to keep a record of the missing values
df['had_missing_salary'] = df['salary_history_2022'].isna()
df['salary_history_2022'].fillna(df['salary_history_2022'].median(), inplace=True)
```

### Cleaning and Transforming Data

```python
# Checking and converting data types
# When you load data, columns don't always come in the right format.
# Numbers might be stored as strings,, dates or plain text

df.dtypes
# Convert joined_date from string to proper datetime format
df['joined_date'] = pd.to_datetime(df['joined_date'])

# Convert a column that has numbers stored as text
# errors = 'coerce' puts NaN where conversion fails (default is 'raise')
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')

# Convert text column with few unique values to categorical
df['employment_type'] = df['employment_type'].astype('category')
```
