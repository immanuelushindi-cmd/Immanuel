### Grouping by Aggregation

#### Groupby
- It works in 3 steps:
    -Split - divides the data into groups
    -Apply - runs a function each group
    -Combine - bring the result together
- for each [category], calculate [something]

```python
df.head()
df.groupby('job_title')['salary_kes'].mean() # average salary per role
df.groupby('city')['name'].count() # Headcount per city
df.groupby('employment_type')['salary_kes'].max() # highest salary per employment type
```

### agg() - Multiple Statistics at Once


```python
df.groupby('job_title')['salary_kes'].agg(['mean', 'median', 'min', 'max', 'count']) 
 
# Different stats for different columns 
df.groupby('job_title').agg({ 
    'salary_kes':     ['mean', 'max'], 
    'experience_yrs': ['mean', 'min'], 
    'age':            'median' 
})
```

### Grouping by Multiple Columns

```python
df.groupby(['city', 'employment_type'])['salary_kes'].mean() 
 
df.groupby(['job_title', 'city']).agg({ 
    'salary_kes':     'mean', 
    'experience_yrs': 'mean' 
}) 
```

### Pivot tables
- Pivot tables do the same  th

```python
pd.pivot_table( 
    df, 
    values='salary_kes', 
    index='job_title', 
    columns='city', 
    aggfunc='mean', 
    fill_value=0 
) 
 
# With row/column totals 
pd.pivot_table( 
df, 
values='salary_kes', 
index='job_title', 
columns='employment_type', 
aggfunc='mean', 
margins=True, 
margins_name='Overall Avg' 
) 
```

### Merging and Combining Datasets

### pd.merge() - Joining Two Tables
- Merging combines 2 Dataframes based on a shared column (like SQL join)

```python
# Second table: department info 
import pandas as pd
departments = pd.DataFrame({ 
    'job_title':  ['Data Analyst', 'Software Engineer', 'Data Scientist', 'ML Engineer', 
'DevOps Engineer'], 
    'department': ['Analytics', 'Engineering', 'Analytics', 'AI/ML', 'Infrastructure'], 
    'budget_kes': [5000000, 8000000, 6000000, 9000000, 7000000] 
}) 

print(departments)
print()
 
# Inner join — only rows where job_title exists in BOTH tables 
merged = pd.merge(df, departments, on='job_title', how='inner') 
print(merged)
```

```python
pd.merge(df, departments, on='job_title', how='inner')  # only matching rows 
pd.merge(df, departments, on='job_title', how='left')   # all left + NaN for missing right 
pd.merge(df, departments, on='job_title', how='right')  # all right + NaN for missing left 
pd.merge(df, departments, on='job_title', how='outer')  # all rows from both tables
```

### pd.concat() - Stacking DataFrames

```python
# Stack rows vertically 
nairobi_staff = df[df['city'] == 'Nairobi'].copy() 
mombasa_staff = df[df['city'] == 'Mombasa'].copy() 
all_staff = pd.concat([nairobi_staff, mombasa_staff], axis=0, ignore_index=True) 
 
# Stack columns horizontally 
extra_info = pd.DataFrame({ 
    'linkedin_url': ['linkedin.com/in/' + name.lower() for name in df['name']] 
}) 
df_extended = pd.concat([df, extra_info], axis=1) 
```
