### Indexing and slicing
- Indexing means grabbing a single element
- Slicing means grabbing a range of elements

#### 1D Indexing and Slicing

```python
# Here, we give one index in the square brackets and it will return the value at that index.
average_daily_sales = np.mean(mock_sales, axis=0)
print(average_daily_sales)

# Access by position
print(average_daily_sales[0]) # first column
print(average_daily_sales[-1]) # last column
print(average_daily_sales[1:3]) # second and third columns

# Slicing: Syntax: array[start:end:step]
# Stop is exclusive - the element at the stop index is not included in the slice
# Step is optional - if not specified, it will default to 1
print(average_daily_sales[:2]) # first two columns
print(average_daily_sales[::2]) # every other column
print(average_daily_sales[::-1]) # reversed
```

### 2D Indexing and  Slicing
- Here we need two indices: [row, column]. The comma separates them.
- A colon by itself means all of them

```python
# Single element: Daily average sales [row, column]
print(mock_sales[0,0]) # first row, first column
print(mock_sales[0,1]) # first row, second column
print(mock_sales[1,0]) # second row, first column
print(mock_sales[-1,-1]) # last row, last column

# Entire row: Daily average sales for each day [row]
print(mock_sales[0,:]) # first row, all columns

# Entire column: Daily average sales for each day [column]
print(mock_sales[:,0]) # all rows, first column

# Subgrid: first 3 days, first 2 products [row, column]
print(mock_sales[:3,0:2]) # first 3 rows, first 2 columns

# Last 3 days, last 2 products [row, column]
print(mock_sales[-3:,-2:]) # last 3 rows, last 2 columns

# Other entire days, first 2 products [row, column]
alternate_days_sales = mock_sales[::2,:] # every other row, all columns
print(alternate_days_sales)
print(alternate_days_sales.shape)


# N/B A MEMORY WARNING:
# Slices are views not copies. If you modify a slice, you modify the original array.
# If you want to create a copy of the slice use the:
# copy() method. if  you want an independent copy
# subgrid_copy = daily_sales[:3,:].copy()
view = mock_sales[:2,:] 
view[0,0] = 9999
print(mock_sales[0,0]) # This will show 9999, indicating that the original array was not modified

# Safer version
mock_sales[0,0] = 5000 # Resetting the value back to the original for demonstration purposes
safe_copy = mock_sales[:2,:].copy() 
safe_copy[0,0] = 9999
print(mock_sales[0,0]) # This will show 5000, indicating the original array was modified
```

### Aggregation and Axis concepts
- Aggregation collapses an array into one or more summary values
- Total, Averages, minimums etc
- The axis parameter controls which direction you collapse

#### Basic Aggregation

```python
# Sum of all values in the entire array
import numpy as np
# 100 DAYS OF SALES DATA ACROSS 4  CATEGORIES
# Rows = days, Columns = categories e.g food, electronics, clothing, furniture
np.random.seed(42)
daily_sales = np.random.normal(loc = 13000, scale = 1500, size = (10,4)).round().astype(int)
print(daily_sales.shape)

# Sum of all values in the entire array
print(daily_sales.sum())

# Product of elements (useful for growth rates)
growth_rate = np.array([1.02, 1.05, 0.98, 1.03])
compound = growth_rate.prod()
print(f"Compound effect: {compound:.4f}")

# Cumulative sum - running total of food sales over 10days
food_cumsum = daily_sales[:,0].cumsum()
print(f"Running Food Revenue:", food_cumsum)

# Cumulative product - chain of multipliers
rates = np.array([1.0, 1.02, 1.05, 0.98, 1.03])
print("Compound after each period:", rates.cumprod().round(4))
```

# Statistical Aggregation

```python
food_sales = daily_sales[:,0] # All 10 days of food
# Mean
print(f"Mean of food sales: {food_sales.mean():.0f}")

# Median (less affected by extreme outliers)
print(f"Median of food sales: {np.median(food_sales):.0f}")

# Standard deviation (how spred out are the values)
print(f"Std Dev: {food_sales.std():.0f}")

# Variance (std squared)
print(f"Variance: {food_sales.var():.0f}")

# Min and max
print(f"Min: {food_sales.min():.0f}")
print(f"Max: {food_sales.max():.0f}")

# argmin and argmax
print(f"Worst day: {food_sales.argmin():.0f}")
print(f"Best day: {food_sales.argmax():.0f}")
```

```python
# Axis Concepts
# Axis controls which direction numpy collapses when it aggregates data

# Axis = 0 - Collapse down through row - one result per column
# What is the average per category across 10 Days
avrg_per_category = daily_sales.mean()
print("Avg per category:", avrg_per_category)

# Axis = 1 - Collapse down through column - one result per row
# What is the total sales per day
total_per_day = daily_sales.sum(axis=1)
print("Total per day:", total_per_day)

# Min per column (worst per category)
worst_per_category = daily_sales.min(axis=0)
print("Worst per category:", worst_per_category)

# Max per row (best per day)
best_category_per_day = daily_sales.max(axis=1)
print("Best category revenue per day:", best_category_per_day)
```

### Boolean Indexing (Filtering)
- Boolean indexing lets you select elements based on condotion. Numpy evaluates the condition, creates a True/False array, and then uses this array to index the array.

```python
food_sales = daily_sales[:,0]

# Which day had food sales above 13000
mask = food_sales > 13000
print(mask)

# Use the mask to get the actual values
high_food_sales = food_sales[mask]
print(high_food_sales)

# Shortcut - combine into one line
print(food_sales[food_sales > 13000])

# How many days exceede the target?
print((food_sales > 13000).sum())
```

## NP.WHERE : Condition Logic on Arrays

```python
# np.where(condition, value if true, value if false)
# Was each day's food sales above 13000? Mark it "High" or "Low"
food_sales = daily_sales[:,0]
food_label = np.where(food_sales > 13000, "High", "Low")
print(food_label)

# Return numeric values
food_flag = np.where(food_sales > 13000, 1, 0)
print(food_flag)

# Keep values above threshold, replace others with the threshold
floor_13k = np.where(food_sales < 13000, 13000, food_sales)
print(floor_13k)
```

### Handling Missing Data(NaN)
- It is a special float value that means "there is no valid number here" - it is the absence of a value.
- Any arithmetic NaN returns NaN, which is why you have to handle it expilicitly.

```python
import numpy as np
# NaN in a float
print(type(np.nan))

# Arithmetic with NaN always returns NaN
print(5+np.nan)

# NaN is never equal to itself
print(np.nan == np.nan)

# The right way to check for NaN
print(np.isnan(np.nan))
```

```python
# Inserting and Detecting NaN
# Simulate a sales dataset with missing values
sales = np.array([
    [12000, 45000, np.nan, 10000],
    [np.nan, np.nan, 20000, 30000],
    [10000, 20000, 30000, np.nan],
    [40000, np.nan, np.nan, 50000]
])

# Detecting NaN - Returns a boolean array
nan_mask =np.isnan(sales)
print(nan_mask)

# Counting missing values
print(f"total missing: {nan_mask.sum()}")

# Counting missing values per column
print("Missing per category: nan_mask.sum(axis=0)")

# Counting missing values per row
print("Missing per say: nan_mask.sum(axis=1)")
```

### Replacing specific values with NaN

```python
# Simulated data where -1 means "sensor not reading"
raw_data = np.array([12000, -1, 11000, -1, 15000, 14000, -1])

# Replace -1 with NaN using np.where
clean = np.where(raw_data == -1, np.nan, raw_data)
print(clean)
```

### Computing with NaN - Use nan-safe functions
- Regular aggregations like .mean()

```python
food_with_gaps = np.array([12000,np.nan, 11000, 15000, np.nan])

# Regular functions propagate NaN values
print(food_with_gaps.mean())
print(food_with_gaps.sum())

# NaN safe versions ignore the NaN cells
print(np.nanmean(food_with_gaps))
print(np.nansum(food_with_gaps))

print(np.nanstd(food_with_gaps))
print(np.nanmin(food_with_gaps))
print(np.nanmax(food_with_gaps))
```
