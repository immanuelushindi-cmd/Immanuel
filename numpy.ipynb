### Indexing and slicing
- Indexing means grabbing a single element
- Slicing means grabbing a range of elements

#### 1D Indexing and Slicing

```python
# Here, we give one index in the square brackets and it will return the value at that index.
average_daily_sales = np.mean(mock_sales, axis=0)
print(average_daily_sales)

# Access by position
print(average_daily_sales[0]) # first column
print(average_daily_sales[-1]) # last column
print(average_daily_sales[1:3]) # second and third columns

# Slicing: Syntax: array[start:end:step]
# Stop is exclusive - the element at the stop index is not included in the slice
# Step is optional - if not specified, it will default to 1
print(average_daily_sales[:2]) # first two columns
print(average_daily_sales[::2]) # every other column
print(average_daily_sales[::-1]) # reversed
```

### 2D Indexing and  Slicing
- Here we need two indices: [row, column]. The comma separates them.
- A colon by itself means all of them

```python
# Single element: Daily average sales [row, column]
print(mock_sales[0,0]) # first row, first column
print(mock_sales[0,1]) # first row, second column
print(mock_sales[1,0]) # second row, first column
print(mock_sales[-1,-1]) # last row, last column

# Entire row: Daily average sales for each day [row]
print(mock_sales[0,:]) # first row, all columns

# Entire column: Daily average sales for each day [column]
print(mock_sales[:,0]) # all rows, first column

# Subgrid: first 3 days, first 2 products [row, column]
print(mock_sales[:3,0:2]) # first 3 rows, first 2 columns

# Last 3 days, last 2 products [row, column]
print(mock_sales[-3:,-2:]) # last 3 rows, last 2 columns

# Other entire days, first 2 products [row, column]
alternate_days_sales = mock_sales[::2,:] # every other row, all columns
print(alternate_days_sales)
print(alternate_days_sales.shape)


# N/B A MEMORY WARNING:
# Slices are views not copies. If you modify a slice, you modify the original array.
# If you want to create a copy of the slice use the:
# copy() method. if  you want an independent copy
# subgrid_copy = daily_sales[:3,:].copy()
view = mock_sales[:2,:] 
view[0,0] = 9999
print(mock_sales[0,0]) # This will show 9999, indicating that the original array was not modified

# Safer version
mock_sales[0,0] = 5000 # Resetting the value back to the original for demonstration purposes
safe_copy = mock_sales[:2,:].copy() 
safe_copy[0,0] = 9999
print(mock_sales[0,0]) # This will show 5000, indicating the original array was modified
```

### Vectorised operations

```python
# Elementwise operations
# These operations are performed on each element of the array index by index.
# 1. Scalar operations: add, subtract, multiply, divide, and power
# Apply 16% VAT on products
products_with_vat = mock_sales * 1.16
print(products_with_vat)
# Subtract a fixed overhead of 1000 KES per product
net_profit = mock_sales - 1000
print(net_profit[0])


```
