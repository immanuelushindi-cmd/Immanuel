### Vectors
- An ordered array of numbers that represents a point in space
- The order matters i.e. 
- Each number in the list is a component (also called elements or entry)
- The dimension of the vector is the number of components. Count them- that is the rule

### Types of Vectors
- Row Vectors- written horizontally
    Row vector a = [2,3,4]
- Column Vectors- written vertically
    Column vector a = [5,6,7]
    

```python
import numpy as np
# A iD numoy array is our basic vector
matatu = np.array([8, 12, 3.5, 120, 0])
print("Shape of matatu is", matatu.shape)

# Column vector (2D array with one coloumn)
col = matatu.reshape(-1, 1)
print("Shape of col is", col.shape)  # (5,1)
```
